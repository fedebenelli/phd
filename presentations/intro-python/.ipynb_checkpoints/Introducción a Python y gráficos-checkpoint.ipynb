### Definición de funciones

Además de las funciones predeterminadas de Python, se pueden definir todas las funciones que se deseen.

````python
def cuadratica(x, a, b, c):
    """Calcula un polinómio cuadrático en el valor x dado
    args:
    x     (float): Valor de x donde calcular
    a,b,c (float): Coeficientes cuadráticos 
    
    returns:
    y     (float): Función cuadrática valuada en x"""
    
    y = a*x**2 + b*x + c
    return y
````

````python
cuadratica(1, 1, 2, 4)
````

````python
X = []
Y = []
a, b, c = 2, 4, 6
for x in range(10):
    
    # Valuar función en x y asignar a variable `y`
    y = cuadratica(x, a, b, c)
    
    # Concatenar valor de x a lista de Xs
    X.append(x)
    
    # Idem y
    Y.append(y)
    
    if y > 100:
        print('El valor de y superó 100!')
        print('y=', y)
        break

print(X)
print(Y)
````

## Formato extra

````python
x = np.linspace(-100, 100, 50)
y = cuadratica(x, 1, 0, 0)

y_min = y.min()
x_min = x[np.where(y == y_min)]

plt.rcParams['font.family'] = 'Trebuchet MS'
plt.rcParams['font.size'] = 12

fig = plt.figure(figsize=(5,5))

plt.plot(x,y, label = '$x^2$')
plt.title('Gráfico de $f(x) = x^2$')
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.xlim(-10,60)
plt.ylim(-50,4000)
plt.annotate(text='Mínimo global',
                 xy=(x_min, y_min),  xycoords='data',
                 xytext=(0.5, 0.5), textcoords='axes fraction',
                 color='orange',
                 arrowprops=dict(arrowstyle='->', color='black'),
                 horizontalalignment='left', verticalalignment='top',
            )

plt.grid(color='lightgray')

plt.legend()
plt.show()
````

## Gráfico de propiedades termodinámicas con GERG-2008

# Ajuste de parámetros

Utilizando el poliniomio Redlich-Kister:

$Q_{ij} = x_i x_j \sum_{k=1}^N A_k(x_i - x_j)^k$

Encontrar el mínimo valor de $N$ que ajusta a los datos experimentales.

Se utiliza la función de SciPy `curve_fit`

````python
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
from pprint import pprint
````

````python
def excess_function(x, *params):
    """Función de exceso genérica para sistema binario
    x: float
        Fracciones molares fase líquida compuesto 1
    params: float, array
        Vector de parámetros a utilizar
    """

    x1 = x
    x2 = 1 - x
    Q_12 = 0
    A_k = params

    for k, A_k in enumerate(params):
        Q_12 = Q_12 + A_k*(x1 - x2)**(k)
    Q_12 = x1*x2*Q_12
    
    return Q_12

def MSE(y, y_hat):
    """Función error medio cuadrado
    y: float, array
        Datos reales.
    y_hat: float, array
        Datos calculados.
    """
    N = len(y)
    MSE = (y - y_hat)**2
    MSE = MSE.sum()/N
    MSE = np.sqrt(MSE)
    return MSE
````

````python
data = pd.read_csv('examples/fitting/excess_volume/data/data.csv')
data.columns = ['x', 've']

data_string = data.set_index('x').to_string()

# Definir valores experimentales
x_exp = 1 - data.x
y_exp = data.ve

# Mostrar datos
plt.figure(figsize=(10,7))
plt.scatter(x_exp, y_exp, marker='x', color='black')
plt.annotate(data_string, xy=(0.75, -0.35))
plt.show()
````

````python
%%time
# Listas de parámetros obtenidos y errores (nos va a servir luego)
parameters = []
errors = []

# Definir valores experimentales
x_exp = 1 - data.x
y_exp = data.ve

n_params = [i for i in range(1, 10)]  # [i for i in range(1, len(y_exp))]

for i in n_params:    
    params_i = curve_fit(excess_function,
                         x_exp,
                         y_exp,
                         p0=[np.ones(i)]
                        )[0]
    parameters.append(params_i)
    
    y_hat = excess_function(x_exp, *params_i)
    
    error = MSE(y_hat, y_exp)
    errors.append(error)
    
    #print(f'Ajuste con N parámetros: {i} | MSE: {error}')
    #print(f'parameters:')
    #pprint(list(parameters[i-1].round(5)))
    #print('-'*79)
````

````python
plt.scatter(n_params, errors)
plt.title('Error vs N parámetros')
plt.show()
````

````python
plt.figure(figsize=(10,7))
plt.scatter(x_exp, y_exp, marker='x', color='black')
x = np.linspace(x_exp.min(), x_exp.max(), 100)

for i in range(2, 6):
    params = parameters[i]
    y_hat = excess_function(x, *params)
    plt.plot(
        x,
        y_hat, 
        label=f'{i+1} | MSE:{errors[i]:.4f}'
    )
    
plt.legend(title='Número de parámetros')
plt.ylim(-0.4, 0.1)
plt.show()
````

# Caso aplicado

Se realizaron extracciones con distintos solventes (orgánico/inorgánico) y medios (ácido/básico)

Se desea generar gráficos de caja con desvíos comparando las extracciones en distintas fases


````python
# Importación de librerías
import pandas as pd              # Lectura de datos tabulados (almacenados en un archivo Excel)
import numpy as np               # Procesamiento de números
import matplotlib.pyplot as plt  # Gráficos
import seaborn as sns            # Complemento para gráficos
````

## Lectura de datos

````python
df=pd.read_excel('examples/statistics-and-plots/data/ensayos.xlsx', index_col=0, engine='openpyxl')
df.head(10)
````

## Limpieza de datos

````python
# Cambiar "---" por NaN y asegurar que se traten los datos numéricos como tal
df = df.replace('---', np.nan)
df[['Masa Total (g)','Rendimiento','Pureza']] = df[['Masa Total (g)','Rendimiento','Pureza']].astype('float')
df.head(10)
````

## Propiedades del gráfico
Se pueden definir propiedades estéticas como dimensiones de figura, tipo de fuente, etc

````python
plt.rcParams['figure.figsize'] = [10, 5]

font = {'family' : 'IBM Plex Sans',
        'weight' : 'normal',
        'size'   : 15}

plt.rc('font', **font)
````

````python
# Definición de DataFrame 'data' que solo contiene datos de interés
#  además, se le eliminan las filas que tengan "Extracto" como fase, ya que no interesan para graficar
data = df.reset_index()[['Ensayo','Fase','Rendimiento','Pureza']]
for i, row in data.iterrows():
    if row['Fase'] == 'Extracto':
        data.drop(i, axis=0, inplace=True)
        
data.head(10)
````

````python
# Gráfico de los rendimientos y pureza según prueba y fase, para los gráficos de pureza se descartan 
#   las filas de Fase Acuosa ya que no corresponden

fig = plt.figure()

def box_graph(df, key):
    sns.boxplot(x='Fase', y=key, data=data, hue='Ensayo')
    plt.ylabel('Rendimiento (%)')
    plt.xticks(rotation=15)
    plt.show()

box_graph(df, 'Pureza')
````

````python
# Redondeo los valores de cantidad total de AC
# Agrupo la data por Prueba y Fase, redondeo a tres decimales y exporto a un Excel

df['Cantidad AC (g)'] = np.round(df['Cantidad AC (g)'].astype('float'), 3)
group = df[['Cantidad AC (g)','Fase','Rendimiento','Pureza']].reset_index().groupby(['Ensayo','Fase'])
group = np.round(group.agg([np.mean,np.std]), decimals=2)
group = group.replace(np.nan, '---')
````

## Yapa: generar tabla de desviaciones

````python
# Del nuevo DataFrame extraigo las columnas que quiero utilizar
columnas_de_interes = [i for (i,y) in group.columns.to_list()[::2]]

# Defino un nuevo Dataframe vacío, con las columnas definidas
#  como las columnas de interés
means = pd.DataFrame(columns = columnas_de_interes)

for column in columnas_de_interes:
    # Al nuevo DataFrame le asigna valores de 
    means[column] = group[(column, 'mean')].astype('str') + ' ± '  +  group[(column, 'std')].astype('str')
    
means #.to_excel('tabla.xlsx')
````
