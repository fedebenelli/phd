### Definición de funciones

Además de las funciones predeterminadas de Python, se pueden definir todas las funciones que se deseen.

````python
def cuadratica(x, a, b, c):
    """Calcula un polinómio cuadrático en el valor x dado
    args:
    x     (float): Valor de x donde calcular
    a,b,c (float): Coeficientes cuadráticos 
    
    returns:
    y     (float): Función cuadrática valuada en x
    """
    
    y = a*x**2 + b*x + c
    return y
````

````python
cuadratica
````

````python
X = []
Y = []
a, b, c = 2, 4, 6
for x in range(10):
    
    # Valuar función en x y asignar a variable `y`
    y = cuadratica(x, a, b, c)
    
    # Concatenar valor de x a lista de Xs
    X.append(x)
    
    # Idem y
    Y.append(y)
    
    if y > 100:
        print('El valor de y superó 100!')
        print('y=', y)
        break

print(X)
print(Y)
````

````python
def densities(fluid, pesp):
    ds = np.linspace(0.1, 20, 10000)

    iso = fluid.isotherm(ds)
     
    roots = np.array(fluid.density_iterator(pesp))
    rho_l, rho_g = roots[:, 0]

    plt.plot(1/ds, iso['p'])
    plt.axvline(1/rho_l, color='gray', ls='--')
    plt.axvline(1/rho_g, color='gray', ls='--')
    plt.axhline(pesp, color='red')
    plt.xlim(0, 1/rho_g*1.2)
    plt.ylim(0, pesp*2)
    plt.show()
    return roots

for p in np.linspace(0.2, 0.4, 5):
    p = p*10**7
    densities(gas, p)
````

## Ajuste de parámetros

Utilizando el poliniomio Redlich-Kister:

$Q_{ij} = x_i x_j \sum_{k=1}^N A_k(x_i - x_j)^k$

Encontrar el mínimo valor de $N$ que ajusta a los datos experimentales.

Se utiliza la función de SciPy `curve_fit`

### Importación de librerías

````python
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
from pprint import pprint
````

### Definición de funciones a utilizar

````python
def excess_function(x, *params):
    """Función de exceso genérica para sistema binario
    x: float
        Fracciones molares fase líquida compuesto 1
    params: float, array
        Vector de parámetros a utilizar
    """

    x1 = x
    x2 = 1 - x
    Q_12 = 0
    A_k = params

    for k, A_k in enumerate(params):
        Q_12 = Q_12 + A_k*(x1 - x2)**(k)
    Q_12 = x1*x2*Q_12
    
    return Q_12

def MSE(y, y_hat):
    """Función error medio cuadrado
    y: float, array
        Datos reales.
    y_hat: float, array
        Datos calculados.
    """
    N = len(y)
    MSE = (y - y_hat)**2
    MSE = MSE.sum()/N
    MSE = np.sqrt(MSE)
    return MSE
````

### Lectura de datos

````python
data = pd.read_csv('examples/fitting/excess_volume/data/data.csv')
data.columns = ['x', 've']

data_string = data.set_index('x').to_string()

# Definir valores experimentales
x_exp = 1- data.x
y_exp = data.ve

# Mostrar datos
plt.figure(figsize=(10,7))
plt.scatter(x_exp, y_exp, marker='x', color='black')
plt.annotate(data_string, xy=(0.75, -0.35))
plt.show()
````

### Ajuste de parámetros

````python
%time
# Listas de parámetros obtenidos y errores (nos va a servir luego)
parameters = []
errors = []

# Definir valores experimentales
x_exp = 1 - data.x
y_exp = data.ve

n_params = [i for i in range(1, len(x_exp))]

for i in n_params:
    params_i = curve_fit(excess_function, x_exp, y_exp, p0=[np.ones(i)])[0]
    parameters.append(params_i)
    
    y_hat = excess_function(x_exp, *params_i)
    
    error = MSE(y_hat, y_exp)
    errors.append(error)
````

````python
plt.scatter(n_params, errors)
plt.title('Error vs N parámetros')
plt.show()
````

````python
# Listas de parámetros obtenidos y errores (nos va a servir luego)
parameters = []
errors = []

# Definir valores experimentales
x_exp = 1 - data.x
y_exp = data.ve

n_params = [i for i in range(1, 14)]

for i in n_params:    
    params_i = curve_fit(excess_function, x_exp, y_exp, p0=[np.ones(i)])[0]
    parameters.append(params_i)
    
    y_hat = excess_function(x_exp, *params_i)
    
    error = MSE(y_hat, y_exp)
    errors.append(error)
    
    if i > 2 and abs(errors[-2] - errors[-1]) < 0.01:
        print(f'Ya no cambia mucho con {i} parámetros..')
        n_params = [i for i in range(1, i+1)]
        break
````

````python
plt.figure(figsize=(10,7))
plt.scatter(x_exp, y_exp, marker='x', color='black')
x = np.linspace(x_exp.min(), x_exp.max(), 100)

for i in range(1, len(parameters)):
    params = parameters[i]
    y_hat = excess_function(x, *params)
    plt.plot(
        x,
        y_hat, 
        label=f'{i+1} | MSE:{errors[i]:.4f}'
    )
    
plt.legend(title='Número de parámetros')
plt.ylim(-0.4, 0.1)
plt.show()
````

````python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def fick_law(L, tau, D):
    
    
    D = D*10**-5

    def diff(C, t):
        dCdt = np.zeros_like(C)

        dCdt[0]    = D * (2*C[1] - 2*C[0])/(dx**2)
        dCdt[1:-1] = D * np.diff(C, 2)/(dx**2)
        dCdt[-1]   = 0

        return dCdt

    tau = 200*60

    nx = 100
    nt = 1000

    X = np.linspace(0, L, nx)
    t = np.linspace(0, tau, nt)

    C0 = 1

    dx = X[1] - X[0]
    dt = t[1] - t[0]

    C0     = np.ones(nx)
    #C0     = C0*porcentajeAC*dens
    C0[-1] = 0

    sol = odeint(diff, C0, t, ml=1, mu=2)

    t = t/60

    with plt.style.context('ggplot'):

        plt.figure(figsize=(6,5))
        plt.plot(t,sol[:,0], label = 'x=0')
        plt.plot(t,sol[:,int(nx*(1/4))], label = 'x=$\\frac{1}{3} L$')
        plt.plot(t,sol[:,int(nx*(2/4))], label = '$\\frac{2}{4} L_c$')
        plt.plot(t,sol[:,int(nx*(3/4))], label = '$\\frac{3}{4} L_c$')
        plt.plot(t,sol[:,-2], label = '$L_c$')

        plt.ylim(0,1)
        plt.axvline(0,ls='--',c='gray')
        plt.legend(loc='upper center', bbox_to_anchor=(0.5, 0.25), shadow=False, ncol=3)

        plt.xlabel('Tiempo (min)')
        plt.ylabel('Concentración (normalizada)')
        plt.show()
````

````python
@interact(K_inf=(0.0, 5.0, 0.1), 
          K_rec=(0.0, 1.0, 0.01), 
          quarantine=False,
          day=(0.0, 10, 1)
         )
def SIR(K_inf, K_rec, quarantine, day):
    def SIR_model(X, t, K_inf, K_rec, quarantine, day):
        S = X[0]
        I = X[1]
        R = X[2]
    
        dSdt = np.zeros_like(S)
        dIdt = np.zeros_like(I)
        dRdt = np.zeros_like(R)
               
        if quarantine and t > day:
            K_inf = K_inf/100

        dSdt = -K_inf*S*I
        dIdt = K_inf*S*I - K_rec*I
        dRdt = K_rec*I

        return [dSdt, dIdt, dRdt]

    S0 = 1
    I0 = 0.01
    R0 = 0
    t = np.linspace(0, 10, 100)
    sol = odeint(SIR_model, [S0,I0,R0], t, args=(K_inf, K_rec, quarantine, day))
    
    S = sol[:, 0]
    I = sol[:, 1]
    R = sol[:, 2]
    
    pico_y = I.max()
    pico_x = t[np.where(I == pico_y)]
        
    with plt.style.context('ggplot'):
        plt.figure(figsize=(10,7))
        plt.title('Covid')
        plt.plot(t, S, label='Suceptibles')
        plt.plot(t, I, label='Infectados')
        plt.plot(t, R, label='recuperados')
        plt.annotate('Pico', xy=(pico_x, pico_y), xycoords='data',
                    xytext=(0.8, 0.9), textcoords='axes fraction',
                    color='black',
                    arrowprops=dict(arrowstyle='->', color='black'),
                    horizontalalignment='left', verticalalignment='top',)
        if quarantine: plt.axvline(day, color='gray', ls='--')
        plt.legend()
````

````python
t = np.linspace(0, 25, 25)/2
````

````python
@interact(t=(150, 350, 10), pesp=(0.5e5, 10e6, 1e5))
def van_der_waals(t, pesp):
    def vdw(a, b, v, t):
        return (R*t)/(v-b) - a/v**2
    
    def vdwd(a, b, v, t):
        return -(R*t)/(v-b)**2 + 2*a/v**2       
    
    def get_roots(a, b, t, pesp):        
        def find_root(pesp, v):
            p = vdw(a, b, v, t)
            it = 1
            while (abs(p - pesp) > pesp/1e6):
                it += 1
                v = v - (vdw(a, b, v, t)-pesp)/vdwd(a, b, v, t)

                p = vdw(a, b, v, t)

                if it > 100:
                    break                
            return v
    
        gas_root = R*t/pesp       
        gas_root = find_root(pesp, gas_root)
        
        liquid_root = (2 + t/tc)/2 * b
        liquid_root = find_root(pesp, liquid_root)
        
        return gas_root, liquid_root


    
        
    R  = 8.31446261815324 
    tc = 304.13
    pc = 7.37e6
    vc = 94.2e-6
    a = (27/64)*(R*tc)**2/pc
    b = (R*tc)/(8*pc)

    #t = tc
    #t = 250   

    

    
    gas_root, liquid_root = get_roots(a, b, t, pesp)
    v = np.linspace(1e-5, 10e-4, 1000)
    p = vdw(a, b, v, t)
    
    plt.axhline(pesp, color='red')
    plt.axvline(gas_root, color='gray', ls='--')
    plt.axvline(liquid_root, color='gray', ls='--')
    
    plt.xlim(1e-5,10e-4)
    plt.ylim(0, 2e7)  
    plt.xticks(rotation=45)
    
    plt.title(fr'Isotherma de $CO_2$ a {t} K')
    plt.ylabel(r'$P (Pa)$')
    plt.xlabel(r'V (m^3)')
    
    plt.plot(v, p)
````
